```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("DesempenhoValorant.xlsx")
df.columns
```

```python
df = pd.read_excel("DesempenhoValorant.xlsx")
plt.title("Desempenho de HeadShot's")
plt.plot(df["Ato"], df["HS %"])
plt.scatter(df["Ato"], df["HS %"], color="k")
plt.xlabel("Episódio e Ato")
plt.ylabel("Taxa de HS (%)")

##Com exceção do primeiro ato pro segundo, a taxa de HS foi aumentando com o tempo, com o pico de taxa de varição do E3_A2 para E3_A3 ocasionado pela troca do mouse!



```

```python
df = pd.read_excel("DesempenhoValorant.xlsx")
plt.title("Desempenho de K/D's")
plt.plot(df["Ato"], df["K/D"], color = "g")
plt.scatter(df["Ato"], df["K/D"], color="k")
plt.xlabel("Episódio e Ato")
plt.ylabel("Razão de K/D")

#Neste gráfico mostra o K/D, isto é, a média de abates/média de mortes.
#Abaixo o gráfico descreve crescimento constante a partir do Episódio 2 e Ato 3
```

```python
df = pd.read_excel("DesempenhoValorant.xlsx")
y = df["WIN%"]
x = df["Ato"]
plt.scatter(x,y, color = "k")
plt.plot(x,y, color = "r")
plt.title("Taxa de vitórias por ato")
plt.xlabel("Episódio e Ato")
plt.ylabel("Win Rate (%)")
## No Episódio o dano causado foi em zero devido a alguma falha no dados da própria Riot.
## No segundo Episódio percebe-se o pico em aproximadamente 60%, seguindo um comportamento oscilatório em seguida, com média tendendo a 55%
```

```python
df = pd.read_excel("DesempenhoValorant.xlsx")
y = df["Damage"]
x = df["Ato"]
plt.scatter(x,y, color = "k")
plt.plot(x,y, color = "r")
plt.title("Dano por ato")
plt.xlabel("Episódio e Ato")
plt.ylabel("Média de dano causado")

#No gráfico do dano, observa-se crescimento a partir do E2_A3 com taxa constante por dois episódios, diminuindo no último a taxa de crescimento.
```
