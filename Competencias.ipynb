```python
nivel = [2,1,1,1,2,2,3]
area = ['English','Python', 'C', 'SQL', 'Excel', 'Estatística', 'HW']
plt.bar(area,nivel, label="1,0 = básico \n2,0 = Intermediário \n3,0 = Avançado", color = 'g')
plt.title("Nível de competências")
plt.xlabel("Eixo de Áreas")
plt.ylabel("Eixo de níveis")
plt.legend()
plt.show()
```
